# Import libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Data

```python
dataset = pd.read_csv('householdtask3.csv')
```

```python
dataset.head(10)
```

# Scatter plot

```python
plt.scatter(dataset['year'], dataset['own'])
plt.title('Scatter plot')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

# Line Chart

```python
plt.plot(dataset['year'])
plt.plot(dataset['own'])
plt.title('Line Chart')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```


# Bar chart

```python
plt.bar(dataset['year'], dataset['own'])
plt.title('Bar chart')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

# Histogram


```python
plt.hist(dataset['income'])
plt.title('Histogram')

plt.show()
```
